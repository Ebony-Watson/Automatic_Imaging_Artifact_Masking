```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,PrecisionRecallDisplay

from utils.unet import DataGenerator, configuration, init_model #see utils for defined model arch etc.
```

# Prepare dataset

```python
TRAIN_PATH = 'D:/Chapter 5/Re-run/artifact_segmentation/train/'
train_ids =  os.listdir('/'.join([TRAIN_PATH, 'images']))

TEST_PATH = 'D:/Chapter 5/Re-run/artifact_segmentation/test/'
test_ids =  os.listdir('/'.join([TEST_PATH, 'images']))

# Parameters
train_params = {'dim': (576,576),
          'batch_size': 4,
          'n_classes': 1,
          'n_channels': 1,
          'shuffle': True,
           'PATH':TRAIN_PATH}

val_params = {'dim': (576,576),
          'batch_size': 4,
          'n_classes': 1,
          'n_channels': 1,
          'shuffle': False,
           'PATH':TRAIN_PATH}

test_params = {'dim': (576,576),
          'batch_size': 1,
          'n_classes': 1,
          'n_channels': 1,
          'shuffle': False,
           'PATH':TEST_PATH}
```

```python
indices_train, indices_val = train_test_split(train_ids, train_size=0.8, test_size=0.2, shuffle=False)
partition={'train':indices_train, 'validation':indices_val,'test':test_ids}

# Generators
training_generator = DataGenerator(partition['train'], **train_params)
validation_generator = DataGenerator(partition['validation'], **val_params)
testing_generator = DataGenerator(partition['test'], **test_params)
```

# Compile and train segmentation model

```python
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import BinaryCrossentropy
from tensorflow.keras.metrics import BinaryIoU
from tensorflow.keras.initializers import HeNormal

#Define model configuration
cn = configuration(
        num_filters_start = 64,
        num_unet_blocks = 5,
        num_filters_end = 1,
        input_width = 576,
        input_height = 576,
        mask_width = 576,
        mask_height = 576,
        input_dim = 1,
        optimizer = Adam(lr=0.002),
        loss = BinaryCrossentropy,
        initializer = HeNormal(),
        batch_size = 4,
        num_epochs = 150,
        metrics = ['BinaryIoU','accuracy'],
        dataset_path = os.path.join(os.getcwd(), 'data'))
```

```python
chkpt = './weights/U-Net_131123_test_2.h5'

config = cn
batch_size = config.get("batch_size")
validation_sub_splits = config.get("validation_sub_splits")
num_epochs = config.get("num_epochs")
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint,ReduceLROnPlateau

# training_callbacks
cp_cb = ModelCheckpoint(filepath = chkpt, monitor='val_loss', verbose=1, save_best_only=True, save_weights_only=True, mode='auto')
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=10, min_lr=0,verbose=1)
es_cb = EarlyStopping(monitor='val_loss', patience=10, verbose=1, mode='auto')
```

```python
# Compute data-dependent variables
train_num_samples = len(indices_train)
val_num_samples = len(indices_val)
steps_per_epoch = train_num_samples // batch_size
val_steps_per_epoch = val_num_samples // batch_size
```

```python
# Initialize model
model = init_model(config = cn)
#model.summary()

model.fit_generator(training_generator,epochs=num_epochs,
                steps_per_epoch=steps_per_epoch,validation_data=validation_generator,
              validation_steps=val_steps_per_epoch, callbacks=[cp_cb, reduce_lr,es_cb],
                verbose=2)
```

```python
print(classification_report(masks.flatten(), preds_x.flatten()))
```

```python
display = PrecisionRecallDisplay.from_predictions(masks.flatten(), preds.flatten(), name="U-Net",color='darkorange')
_ = display.ax_.set_title("Precision-Recall curve for Artifact Segmentation")
#plt.savefig(f'./results/PRCurve_UNET_testdataset.svg')
plt.legend(frameon=False)
```

### Image-wise performance of model

```python
keys = ['F1 Score', 'IOU','Recall','Precision']
eval_dict = {k: [] for k in keys}

for man, unet in zip(masks, preds_x):
    f1 = metrics.f1_score(man.flatten(), unet.flatten())
    eval_dict['F1 Score'].append(f1)
    iou = metrics.jaccard_score(man.flatten(), unet.flatten())
    eval_dict['IOU'].append(iou)
    recall = metrics.recall_score(man.flatten(), unet.flatten())
    eval_dict['Recall'].append(recall)
    precision = metrics.precision_score(man.flatten(), unet.flatten())
    eval_dict['Precision'].append(precision)
```

```python
dat = pd.DataFrame.from_dict(eval_dict)
print(dat.median())
```

```python
plt.figure(1)
sns.kdeplot(data=dat.melt(), x="value",hue='variable',hue_order=[ 'Precision','Recall', 'F1 Score','IOU'], palette=['darkorange','gray','teal', 'purple',])
plt.ylabel('Density')
plt.xlabel('Image-wise Performance')
plt.gca().spines[['top', 'right']].set_visible(False)
#plt.savefig(f'./results/UNet-ImageWise_testdataset.svg', dpi=300, transparent=True,bbox_inches='tight')
plt.show()
```
